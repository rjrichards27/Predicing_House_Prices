## Data Cleaning

```python
# finding relationship between price and variables
price_corr = all_data2.corr()["price"]
price_corr.abs().sort_values(ascending=False)[1:]
```

```python
all_data2.info()
```

```python
# transfrom non-int variables to int
all_data2[['bathrooms', 'bedrooms', 'garageSpaces']] = all_data2[['bathrooms', 'bedrooms', 'garageSpaces']].astype('int64')
```

```python
# drop the 'livingAreaValue' column, it's almost the same as 'livingArea'
#          'county', we have 'countId'
all_data2 = all_data2.drop(['livingAreaValue', 'county'], axis=1)
```

```python
# drop the 'buildingArea' column, because
# 1. most of the building area is 0
print(all_data2.buildingArea[all_data2['buildingArea']==0].count(), '\n')

# 2. except for rows that are 0, the values of buildingArea and livingArea are exactly the same
temp = all_data2[['livingArea','buildingArea', 'lotAreaUnits']]

temp = temp[temp['buildingArea']!=0]
print(temp.livingArea.describe(), '\n')
print(temp.buildingArea.describe())

all_data2 = all_data2.drop('buildingArea', axis=1)
```

```python
# remove outliers - year
all_data2.yearBuilt.unique()
```

```python
years_out = (all_data2.yearBuilt == 0) | (all_data2.yearBuilt == 9999)

all_data2 = all_data2[~years_out]
all_data2.shape
```

```python
# remove outliers - bathrooms
all_data2.bathrooms.unique()
```

```python
bath_out = (all_data2.bathrooms == 0) | (all_data2.bathrooms > 20)

all_data2 = all_data2[~bath_out]
all_data2.shape
```

```python
# remove outliers - bedrooms
all_data2.bedrooms.unique()
```

```python
bed_out = (all_data2.bedrooms == 0) | (all_data2.bedrooms > 20)

all_data2 = all_data2[~bed_out]
all_data2.shape
```
