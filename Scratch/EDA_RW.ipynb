```python
import pandas as pd
import numpy as np
from pathlib import Path
import zipfile


#importing the california data
PARENT = "Predicting_House_Prices"
path = Path(PARENT).parent / "../Data/RealEstate_Georgia.csv"
ga = pd.read_csv(path)
ga.head()
```

```python
#importing the georgia data
path2 = Path(PARENT).parent / "../Data/RealEstate_California.csv.zip"
zf = zipfile.ZipFile(path2)
ca = pd.read_csv(zf.open("RealEstate_California.csv"))
ca.head()
```

```python
#combining the two datasets
all_data = pd.concat([ca, ga])
all_data
```

```python
#looking at the columns in the georgia data
print(all_data.columns)
```

```python
all_data.loc[all_data.hasBadGeocode==1,]
```

```python
all_data['livingArea_c']=all_data["livingArea"] - all_data.livingArea.mean()
all_data['price_c']=all_data['price'] - all_data.price.mean()

all_data.loc[all_data.livingArea_c >= all_data.livingArea_c.quantile(q=0.99), 'livingArea_c'] = np.NaN
all_data.loc[all_data.price_c >= all_data.price_c.quantile(q=0.99), 'price_c'] = np.NaN


plt.figure(figsize=(8, 8))
plt.scatter(all_data.livingArea_c, all_data.price_c)
plt.xlabel("Living Area")
plt.ylabel("Price")
plt.title("House Prices vs Living Area")
plt.show()
```
